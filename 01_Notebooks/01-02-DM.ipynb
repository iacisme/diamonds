# DATA TRANSFORMATION

Data transformation is the process of augmenting a data set with new features, created by performing transformations of the existing ones. It can also include scaling features. Different features may have different ranges. These ranges may differ by several orders of magnitude. 

## NEW FEATURES

The new features that will be added to this dataset are:

1. [X] **Carat Weight:** Categorizing diamonds by carat weight into small, medium, and large diamonds.
2. [X] **Color gradings:** Create a categories for color gradings
3. [X] **Price:** These prices are from 7 years ago, what would be the inflation prices to buy them in 2024?

### CARAT WEIGHT CATEGORIZING

The `Diamonds Type` feature will be created to group carat weight by Diamond type according to the following criteria:

* **Small Diamond:** for diamonds up to 0.29 carat weight
* **Medium Diamond:** for diamonds from 0.3 to 0.99 carat weight
* **Large Diamond:** for diamonds > 1.0 carat weight

These categories will help sort diamonds by common characteristics, to help expedite pattern recognition.

```python
# Create a new column named 'Diamond Type' based on carat weight ranges
diamonds_df['Diamond Type'] = pd.cut(
    diamonds_df['Carat'], 
    labels =  ["Small Diamond", "Medium Diamond", "Large Diamond"],
    bins = [0, 0.3, 1, float('inf')],
)

# Verify the transformation was performed correctly
diamonds_df
```

### COLOUR GRADING

The diamond colour grading scale actually measures the lack of colour. The less colour found in a diamond, the more valuable the diamond, according to this scale:

1. **Colourless Diamonds:** Diamonds with a colour grading from D to F.
2. **Near Colourless Diamonds:** Diamonds with a colour grading G to J.
3. **Faint:** Diamonds with a colour grading from K to N.
4. **Very Light:** Diamonds with a colour grading from O to R.
5. **Light:** Diamonds with a colour grading from S to Z.

The following scripts will create this new feature.

```python
# Create a list of color grades
colorGrade_list = diamonds_df['Color']
```

```python
# Function to categorize a color grading
def categorizeColors(colorGrade):
    # Criteria of grading a diamond
    if colorGrade in ['D', 'E', 'F']:
        return 'Colorless'
    elif colorGrade in ['G', 'H', 'I', 'J']:
        return 'Near Colorless'
    elif colorGrade in ['K', 'L', 'M', 'N']:
        return 'Faint'
    elif coloGrade in ['O', 'P', 'Q', 'R']:
        return 'Very Light'
    elif colorGrade in ['S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z']:
        return 'Light'
```

```python
# Create a blank list to store the new categories
colorCategories_list = []
```

```python
# Create the new feature
for colorGrade in colorGrade_list:
    grade = categorizeColors(colorGrade)
    colorCategories_list.append(grade)
```

```python
# Confirm the new list is the same length as original list (True)
len(colorGrade_list) == len(colorCategories_list)
```

```python
# Add the column to the dataset
diamonds_df['Color Grades'] = colorCategories_list

# Verify the transformation operation
diamonds_df
```

### PRICE

The prices in this dataset are from 2017. A new price column is going to be created, to adjust the prices for inflation.

The inflation rate that will be used in this project is: `8.60%`. It is based on the inflation rates found in the following websites:

* https://www.in2013dollars.com/Jewelry/price-inflation/2017-to-2024?amount=326 (8.34%)
* https://calcinflation.com/diamond-jewelry-inflation-calculator/ (13.26%)
* https://calcinflation.com/ (8.57%)

The following scripts will create inflation adjusted prices:

```python
# Inflation rate (as a decimal)
inflationRate = 8.60 / 100
```

```python
# Create a new column of 2024 diamond pricing based on an 8.60% inflation rate from 2017 to 2024 
diamonds_df['Price(2024)'] = round(diamonds_df['Price'] * (1 + inflationRate), 2)

diamonds_df
```
